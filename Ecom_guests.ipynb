```python
import pandas as pd

# Завантаження данних з файлу Excel в DataFrame
df = pd.read_excel('Тестове завдання Data Analyst CVM_1.xlsx')

# Перші декілька рядків таблиці
print(df.head())
```

```python
# Отримання статистичної інформації про числові колонки
print(df.describe())

# Вивчення розподілу віку гостей
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['age'], bins=10, kde=True)
plt.title('Розподіл віку гостей')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()

# Вивчення кореляції між числовими атрибутами
correlation = df.corr()
print(correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Встановлення стилю графіків
sns.set(style="whitegrid")

# Створення фігури та осей для побудови графіків
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Список атрибутів для побудови гістограм
attributes = ['kol_vo_ch_online', 'avg_ch_online', 'kol_vo_ch_offline', 'avg_ch_offline',
              'have_mob_app', 'kol_vo_session_mob_app']

# Побудова гістограм для кожного атрибута
for i, attr in enumerate(attributes):
    sns.histplot(df[attr], ax=axes[i//3, i%3], kde=True)
    axes[i//3, i%3].set_title(f'Distribution of {attr}')

plt.tight_layout()
plt.show()
```
