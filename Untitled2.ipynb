```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантажте або визначте ваш набір даних
test_data = pd.read_csv('ab_test_data.csv')

# Перетворіть колонку 'timestamp' на тип даних дати
test_data['timestamp'] = pd.to_datetime(test_data['timestamp'])

# Розділіть дані за групами
data_a = test_data[test_data['test_group'] == 'a']
data_b = test_data[test_data['test_group'] == 'b']

# Агрегуйте дані за день
data_a_daily = data_a.groupby(data_a['timestamp'].dt.date)['conversion'].mean().reset_index()
data_b_daily = data_b.groupby(data_b['timestamp'].dt.date)['conversion'].mean().reset_index()

# Побудуйте графік
plt.figure(figsize=(10, 6))
sns.lineplot(x='timestamp', y='conversion', data=data_a_daily, label='A')
sns.lineplot(x='timestamp', y='conversion', data=data_b_daily, label='B')
plt.title('Conversion Rate Over Time (Daily Aggregation)')
plt.xlabel('Time')
plt.ylabel('Conversion Rate')
plt.legend()
plt.show()
```
